```python
with open(file) as f:
    input_str = f.read()[:-1]

input_str[:100], input_str[-1]
```

```python
input_list = input_str.split("Monkey")[1:]
input_dict = { inp[1]:inp[6:] for inp in input_list} #stripping trailing spaces
input_dict
```

```python
def parse_monkey(monkey_str):
    pattern = r"Starting items: (.*?)\n  Operation: new = old (\+|\*) (.*?)\n  Test: divisible by (.*?)\n    If true: throw to monkey (\d+)\n    If false: throw to monkey (\d+)"
    res = re.search(pattern, monkey_str)
    
    d = {}
    
    pattern = "(\d+)"
    try:
        d['starting_items'] = [int(i.group(0)) for i in re.finditer(pattern, res.group(1))]
    except:
        print(res.group(1))
    
    d['operator'] = res.group(2) # + or *
    d['operand'] = res.group(3) # "old" or int
    d['test_div_by'] = int(res.group(4))
    d['targets'] = {True: int(res.group(5)), False: int(res.group(6))}
    
    return d
    # self.starting_items, self.op, self.test, self.targets

ex_str = "Starting items: 79\n  Operation: new = old * 19\n  Test: divisible by 23\n    If true: throw to monkey 2\n    If false: throw to monkey 3\n\n"
print(parse_monkey(ex_str))
```

```python
class Monkey:
    
    @classmethod
    def set_class_variable_all_monkeys(cls, value):
        cls.all_monkeys = value
    
    def __init__(self, monkey_num, monkey_str, divider=None):
        self.monkey_num = monkey_num
        self.monkey_dict = parse_monkey(monkey_str)
        self.operation_dict = {"*": operator.mul,"+": operator.add}
        self.inspect_count = 0
        self.divider = divider
            
     
    def examine(self):
        #print(f"\nüêµ this is monkey {self.monkey_num} examining my list:{self.monkey_dict['starting_items']}")
        vals_to_treat = self.monkey_dict['starting_items'].copy()
        for item_val in vals_to_treat:
            #print(item_val)
            self.treat(item_val)
            
    def treat(self, item_val):
        if self.monkey_dict['operand'] == 'old':
            operand = item_val
        else: operand = int(self.monkey_dict['operand'])
        new_val = self.operation_dict[self.monkey_dict['operator']](item_val, operand)
        if self.divider:
            new_val = floor(new_val/self.divider)
        next_monkey = self.monkey_dict['targets'][new_val % self.monkey_dict['test_div_by'] == 0]
        #print(f"I examined item {item_val} and will throw it to monkey {next_monkey}")

        self.pass_on(new_val, next_monkey)
            
    def pass_on(self, new_val, next_monkey):
        self.inspect_count += 1
        #print(f'my vals before pop:{self.monkey_dict["starting_items"]}.')
        self.monkey_dict["starting_items"].pop(0)
        #print(f'my vals after pop:{self.monkey_dict["starting_items"]}.')
        #print(f'their vals before append:{self.all_monkeys[next_monkey].monkey_dict["starting_items"]}.')
        self.all_monkeys[next_monkey].monkey_dict["starting_items"].append(new_val)
        #print(f'their vals after append:{self.all_monkeys[next_monkey].monkey_dict["starting_items"]}.\n')
        
            

        
```

```python
monkeys_dict = {int(monkey[0]) : Monkey(int(monkey[0]), monkey[1], divider=3) for monkey in input_dict.items()}

monkeys_dict[0].set_class_variable_all_monkeys(monkeys_dict)

for r in range(20):
    print(f"ü•Åü•Åü•Å ROUND {r}")
    for i in range(len(monkeys_dict)):
        monkeys_dict[i].examine()
    
```

```python
# --- part 2 -- unmanageable in same way as 1
dividers = [m.monkey_dict['test_div_by'] for m in monkeys_dict.values()]
dividers.sort()
dividers # all prime --> Chinese remainder theorem --> rewrite 
maxi_divider = prod(dividers)
```

```python
class Monkey2:
    
    @classmethod
    def set_class_variable_all_monkeys(cls, value):
        cls.all_monkeys = value
    
    def __init__(self, monkey_num, monkey_str, divider=None):
        self.monkey_num = monkey_num
        self.monkey_dict = parse_monkey(monkey_str)
        self.operation_dict = {"*": operator.mul,"+": operator.add}
        self.inspect_count = 0
        self.divider = divider
        self.maxi_divider=1
            
     
    def examine(self):
        #print(f"\nüêµ this is monkey {self.monkey_num} examining my list:{self.monkey_dict['starting_items']}")
        vals_to_treat = self.monkey_dict['starting_items'].copy()
        for item_val in vals_to_treat:
            #print(item_val)
            self.treat(item_val)
            
    def treat(self, item_val):
        if self.monkey_dict['operand'] == 'old':
            operand = item_val
        else: operand = int(self.monkey_dict['operand'])
        new_val = self.operation_dict[self.monkey_dict['operator']](item_val, operand)
        if self.divider:
            new_val = floor(new_val/self.divider)
        else:
            new_val = new_val % maxi_divider 
            #(modulo the product of all prime dividers shouldn't change next test)
            # few will yield something else than themselves but that may be enough to decrease the fever
        next_monkey = self.monkey_dict['targets'][new_val % self.monkey_dict['test_div_by'] == 0]
        #print(f"I examined item {item_val} and will throw it to monkey {next_monkey}")

        self.pass_on(new_val, next_monkey)
            
    def pass_on(self, new_val, next_monkey):
        self.inspect_count += 1
        #print(f'my vals before pop:{self.monkey_dict["starting_items"]}.')
        self.monkey_dict["starting_items"].pop(0)
        #print(f'my vals after pop:{self.monkey_dict["starting_items"]}.')
        #print(f'their vals before append:{self.all_monkeys[next_monkey].monkey_dict["starting_items"]}.')
        self.all_monkeys[next_monkey].monkey_dict["starting_items"].append(new_val)
        #print(f'their vals after append:{self.all_monkeys[next_monkey].monkey_dict["starting_items"]}.\n')
        
            

        
```

```python
monkeys_dict = {int(monkey[0]) : Monkey2(int(monkey[0]), monkey[1], divider=None) for monkey in input_dict.items()}

monkeys_dict[0].set_class_variable_all_monkeys(monkeys_dict)
monkeys_dict[0].maxi_divider = maxi_divider
```
